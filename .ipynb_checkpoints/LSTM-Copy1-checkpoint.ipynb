```python
import numpy as np
from pathlib import Path
import glob

import matplotlib.pyplot as plt
import random
from scipy.signal import savgol_filter

import pickle

from scipy.signal import resample
import pandas as pd
path='/home/predator/Documents/redpitaya_ws/LSTM/s20240110'
```

```python
def resampling(signal,f,f2):
    signal_resampled = resample(signal, int(len(signal) * f2 / f))
    return (signal_resampled)

def load_data(fullname):
    with open(fullname, 'rb') as fd:
          data = pickle.load(fd)
          fd.close()
    return data

def scaledata(data):
    settings = data['settings']
    trace = np.frombuffer(data['data'], dtype='>u2')
    gain = float(settings.split("YINC:")[1].split("\n")[0])
    offset = float(settings.split("YOR:")[1].split("\n")[0])
    trace = trace * gain - offset
    return(trace)

def savitzky(data):
    window_length = 4001 if 4000 % 2 == 0 else 4000
    return savgol_filter(data, window_length, 3)


def disp_treat(disp):
    disp=disp[:(len(disp)-len(disp)%256)]
    X=len(disp)//256
    disp=np.reshape(disp,(X,256))
    train_displacement=[disp[i,255]-disp[i,0] for i in range (X)]
    train_displacement=np.reshape(train_displacement,(X,1))
    return train_displacement
```

```python
n_pts=256
i='126'
fsamp=256e3
print("resampling frequency:",fsamp, 'Hz')
```

```python
liste_ch1 = sorted(glob.glob(path+'/s20240110_*_ch{}.pickle'.format(1)))
liste_ch3 = sorted(glob.glob(path+'/s20240110_*_ch{}.pickle'.format(3)))
liste_ch4 = sorted(glob.glob(path+'/s20240110_*_ch{}.pickle'.format(4)))
```

```python
file3 = [f for f in liste_ch4 if f"s20240110_{i}_rtb_ch4.pickle" in f][0]
data3=scaledata(load_data(file3))
#data3=resampling(data3,f=1/sampling_time,f2=fsamp)
data3=data3[:(len(data3)//256)*256]
```

```python
def create_sequences(signal, velocity, seq_length=256):
    X, y = [], []
    
    # Ensure we do not go beyond the velocity array
    max_length = min(len(signal) - seq_length, len(velocity))  

    for i in range(max_length):
        X.append(signal[i:i + seq_length])  # 256 points as input
        y.append(velocity[i])  # 1 velocity value as output
    
    X = np.array(X)
    y = np.array(y).reshape(-1, 1)  # Reshape target to (samples, 1)
    
    return X, y

# Extract features (input) and labels (output)
velocity_data = disp_treat(savitzky(data2))  # Precompute velocity
X, y = create_sequences(data1, velocity_data, seq_length=256)

print("Input shape:", X.shape)  # Should be (num_samples, 256)
print("Output shape:", y.shape)  # Should be (num_samples, 1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)

scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

# 80% Training, 20% Testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

print("Training set:", X_train.shape, y_train.shape)
print("Testing set:", X_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Inverse transform
y_train_original = scaler_y.inverse_transform(y_train)
y_test_original = scaler_y.inverse_transform(y_test)

# Plot
plt.figure(figsize=(12, 5))
plt.plot(xbis,y_train_original[:100], label="Training Data", color='blue', linestyle='solid', alpha=0.7)
plt.plot(y_test_original[:100], label="Testing Data", color='red', linestyle='dashed', alpha=0.7)

plt.legend()
plt.xlabel("Time (ms)")
plt.ylabel("Velocity (original scale)")
plt.title("Training vs Testing Data (original scale)")
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# Reshape X to (samples, timesteps, features) for LSTM
X_train = X_train.reshape(X_train.shape[0], 256, 1)  # Single feature (interferometric signal)
X_test = X_test.reshape(X_test.shape[0], 256, 1)

# Define LSTM model
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(256, 1)),
    LSTM(32, return_sequences=False),
    Dense(16, activation='relu'),
    Dense(1)  # Single output (velocity)
])

# Compile model
model.compile(loss='mse', optimizer=Adam(learning_rate=0.001), metrics=['mae'])

# Early stopping
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train model
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=16,  # Reduce from 32 to 16
    callbacks=[early_stop]
)

```

```python
# Evaluate on test set
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss:.4f}, Test MAE: {test_mae:.4f}")

# Make predictions
y_pred = model.predict(X_test)

# Convert back to original scale
y_pred_original = scaler_y.inverse_transform(y_pred)
y_test_original = scaler_y.inverse_transform(y_test)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(y_test_original[:100], label="True Velocity", color='blue')
plt.plot(y_pred_original[:100], label="Predicted Velocity", color='red', linestyle='dashed')
plt.legend()

# Ensure xlabel is correctly set
plt.xlabel("Sample Index")
plt.ylabel("Velocity")
plt.title("LSTM Prediction vs True Velocity")

plt.show()
```
