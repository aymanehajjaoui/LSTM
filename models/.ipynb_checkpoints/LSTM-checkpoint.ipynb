# Imports

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import load_model

from scipy.signal import savgol_filter, resample
import pickle, glob, os

from scipy.stats import pearsonr
import seaborn as sns

import pandas as pd
```

# Parameters

```python
date_id = "20250319"
record_id = "300"
base_path = f"/home/predator/Documents/redpitaya_ws/LSTM/s{date_id}"
channels = [1, 2, 3, 4]
segment_length = 256
velocity_channel = "ch3"
```

# Functions

```python
def load_data(fullname):
    with open(fullname, 'rb') as fd:
        return pickle.load(fd)

def scaledata(data):
    settings = data['settings']
    trace = np.frombuffer(data['data'], dtype='>u2')
    gain = float(settings.split("YINC:")[1].split("\n")[0])
    offset = float(settings.split("YOR:")[1].split("\n")[0])
    return trace * gain - offset

def segment_velocity(signal, segment_length=256):
    signal = signal[:(len(signal) // segment_length) * segment_length]
    reshaped = signal.reshape((-1, segment_length))
    velocity = np.array([seg[-1] - seg[0] for seg in reshaped])
    return velocity

def savitzky(data):
    window_length = 4001 if 4000 % 2 == 0 else 4000
    return savgol_filter(data, window_length, 3)
```

# Load data

```python
data_dict = {}
sampling_time = None

for ch in channels:
    search_pattern = os.path.join(base_path, f"s{date_id}_{record_id}_rtb_ch{ch}.pickle")
    file = glob.glob(search_pattern)[0]
    data_raw = load_data(file)
    signal = scaledata(data_raw)
    signal = signal[:(len(signal) // segment_length) * segment_length]
    if ch == 1:
        sampling_time = float(data_raw['settings'].split("YINC:")[1].split("\n")[0])
    data_dict[f"ch{ch}"] = signal

x = np.arange(0, len(data_dict['ch1']) * sampling_time, sampling_time)
```

# Velocity Calculation

```python
raw_signal = data_dict[velocity_channel]
filtered_signal = savitzky(raw_signal)

velocity_raw = segment_velocity(raw_signal, segment_length)
velocity_filtered = segment_velocity(filtered_signal, segment_length)
xbis = np.linspace(x[0], x[-1], len(velocity_raw))
```

# Plot All

```python
titles = {
    "ch1": "(Ch1)",
    "ch2": "(Ch2)",
    "ch3": "(Ch3)",
    "ch4": "(Ch4)"
}

fig, axs = plt.subplots(len(channels) + 1, 1, figsize=(12, 12), sharex=True)

for idx, ch in enumerate(channels):
    ax = axs[idx]
    y = data_dict[f"ch{ch}"]
    ax.plot(x, y, linewidth=0.7)
    ax.set_ylabel(titles.get(f"ch{ch}", f"Channel {ch}"))

    for i in range(0, len(x), segment_length):
        color = 'lightblue' if (i // segment_length) % 2 == 0 else 'lightgreen'
        ax.axvspan(x[i], x[min(i + segment_length - 1, len(x) - 1)], color=color, alpha=0.3)

# Velocity plot (5th plot)
ax_v = axs[-1]
ax_v.plot(xbis, velocity_raw, label="Raw velocity", color="tab:blue")
ax_v.plot(xbis, velocity_filtered, label="Filtered velocity", color="tab:orange")
ax_v.set_ylabel("Velocity")
ax_v.set_xlabel("Time (s)")
ax_v.legend()

for i in range(len(xbis)-1):
    color = 'lightblue' if i % 2 == 0 else 'lightgreen'
    ax_v.axvspan(xbis[i], xbis[i+1], color=color, alpha=0.2)

fig.suptitle("Raw Experimental Signals + Velocity (5 plots)")
fig.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

# Prints

```python
print("Sampling frequency:", 1/sampling_time, "Hz")
print("Total segments:", len(data_dict['ch1']) // segment_length)
print("Full length (s):", max(x))
```

# Preparing Dataset

```python
y = velocity_filtered.reshape(-1, 1)

# Reshape X for LSTM input: (samples, timesteps, features)
X = data_dict['ch1'][:len(y) * segment_length].reshape(-1, segment_length, 1)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Reconstruct the full-length interferometric signal
X_continuous = X.squeeze().flatten()  # Shape: (num_samples * 256,)
y_flat = y.squeeze()

# Rebuild time axis for full signal
t_signal = np.arange(len(X_continuous)) * sampling_time

# Rebuild time axis for velocity (1 point per 256 samples)
t_velocity = np.linspace(t_signal[0], t_signal[-1], len(y_flat))

# Plot both
plt.figure(figsize=(14, 6))

# --- Full signal (input to LSTM) ---
plt.subplot(2, 1, 1)
plt.plot(t_signal, X_continuous, label="Interferometric Signal (Ch1)")
plt.title("Full Interferometric Signal (Input to LSTM)")
plt.ylabel("Amplitude")
plt.legend()

# --- Corresponding velocity ---
plt.subplot(2, 1, 2)
plt.plot(t_velocity, y_flat, label="Target Velocity", color='orange')
plt.title("Velocity Target (1 per 256 samples)")
plt.xlabel("Time (s)")
plt.ylabel("Velocity")
plt.legend()

plt.tight_layout()
plt.show()
```

# Split Data into Training and Testing Sets

```python
# Split the data (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

```python
# Flatten X and y to plot continuously
X_train_cont = X_train.reshape(-1)
X_test_cont = X_test.reshape(-1)
y_train_cont = y_train.flatten()
y_test_cont = y_test.flatten()

# Time axes
t_X_train = np.arange(len(X_train_cont))
t_X_test = np.arange(len(X_test_cont))
t_y_train = np.arange(len(y_train_cont))
t_y_test = np.arange(len(y_test_cont))

# Plot
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.plot(t_X_train, X_train_cont, color='blue')
plt.title("X_train (Continuous)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(t_y_train, y_train_cont, color='red')
plt.title("y_train (Continuous)")
plt.ylabel("Velocity")
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(t_X_test, X_test_cont, color='green')
plt.title("X_test (Continuous)")
plt.xlabel("Time step")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(t_y_test, y_test_cont, color='orange')
plt.title("y_test (Continuous)")
plt.xlabel("Sample index")
plt.ylabel("Velocity")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Save training and test datasets as CSV

```python
# Output folder
output_dir = f"datasets/s{date_id}_{record_id}_dataset_256"
os.makedirs(output_dir, exist_ok=True)

# Save data
pd.DataFrame(X_train_cont).to_csv(f"{output_dir}/x_train.csv", index=False, header=["amplitude"])
pd.DataFrame(X_test_cont).to_csv(f"{output_dir}/x_test.csv", index=False, header=["amplitude"])
pd.DataFrame(y_train_cont).to_csv(f"{output_dir}/y_train.csv", index=False, header=["velocity"])
pd.DataFrame(y_test_cont).to_csv(f"{output_dir}/y_test.csv", index=False, header=["velocity"])

print("CSV files saved as:")
print(f"- {output_dir}/x_train.csv")
print(f"- {output_dir}/x_test.csv")
print(f"- {output_dir}/y_train.csv")
print(f"- {output_dir}/y_test.csv")
```

# Resample to 48

```python
target_length = 48

# Resample each segment in X
X_resampled = np.array([
    resample(segment.squeeze(), target_length) for segment in X
]).reshape(-1, target_length, 1)

print("Resampled X shape:", X_resampled.shape)

# Flatten X for CSV
X_train_res, X_test_res, y_train_res, y_test_res = train_test_split(
    X_resampled, y, test_size=0.2, shuffle=False
)

X_train_res_cont = X_train_res.reshape(-1)
X_test_res_cont = X_test_res.reshape(-1)
y_train_res_cont = y_train_res.flatten()
y_test_res_cont = y_test_res.flatten()

# Output folder for resampled dataset
resampled_output_dir = f"datasets/s{date_id}_{record_id}_resampled_48"
os.makedirs(resampled_output_dir, exist_ok=True)

# Save resampled data
pd.DataFrame(X_train_res_cont).to_csv(f"{resampled_output_dir}/x_train.csv", index=False, header=["amplitude"])
pd.DataFrame(X_test_res_cont).to_csv(f"{resampled_output_dir}/x_test.csv", index=False, header=["amplitude"])
pd.DataFrame(y_train_res_cont).to_csv(f"{resampled_output_dir}/y_train.csv", index=False, header=["velocity"])
pd.DataFrame(y_test_res_cont).to_csv(f"{resampled_output_dir}/y_test.csv", index=False, header=["velocity"])

print("Resampled CSV files saved as:")
print(f"- {resampled_output_dir}/x_train.csv")
print(f"- {resampled_output_dir}/x_test.csv")
print(f"- {resampled_output_dir}/y_train.csv")
print(f"- {resampled_output_dir}/y_test.csv")
```

```python
print("Resampled training set shape:", X_train_res.shape, y_train_res.shape)
print("Resampled testing set shape:", X_test_res.shape, y_test_res.shape)
```

# Normalize the Data

```python
# Standardize X and y separately
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train.reshape(-1, segment_length)).reshape(X_train.shape)
X_test_scaled = scaler_X.transform(X_test.reshape(-1, segment_length)).reshape(X_test.shape)

scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train)
y_test_scaled = scaler_y.transform(y_test)

print("X_train_scaled shape:", X_train_scaled.shape)
print("y_train_scaled shape:", y_train_scaled.shape)
```

```python
X_train_cont = X_train_scaled.reshape(-1)
X_test_cont = X_test_scaled.reshape(-1)
y_train_cont = y_train_scaled.flatten()
y_test_cont = y_test_scaled.flatten()

# Time axes
t_X_train = np.arange(len(X_train_cont))
t_X_test = np.arange(len(X_test_cont))
t_y_train = np.arange(len(y_train_cont))
t_y_test = np.arange(len(y_test_cont))

# Plot
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.plot(t_X_train, X_train_cont, color='blue')
plt.title("X_train_scaled (Continuous)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(t_y_train, y_train_cont, color='red')
plt.title("y_train_scaled (Continuous)")
plt.ylabel("Velocity (scaled)")
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(t_X_test, X_test_cont, color='green')
plt.title("X_test_scaled (Continuous)")
plt.xlabel("Time step")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(t_y_test, y_test_cont, color='orange')
plt.title("y_test_scaled (Continuous)")
plt.xlabel("Sample index")
plt.ylabel("Velocity (scaled)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Build the LSTM Model

```python
# Define the LSTM model
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(segment_length, 1)),
    LSTM(32, return_sequences=False),
    Dense(16, activation='relu'),
    Dense(1)
])

# Compile model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])

# Summary
model.summary()
```

# Train the Model

```python
# Early stopping to prevent overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train the model
history = model.fit(
    X_train_scaled, y_train_scaled,
    validation_data=(X_test_scaled, y_test_scaled),
    epochs=30,
    batch_size=32,
    callbacks=[early_stop]
)
```

# Plot Training History

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss (MSE)")
plt.title("Loss Curve")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Train MAE')
plt.plot(history.history['val_mae'], label='Val MAE')
plt.xlabel("Epochs")
plt.ylabel("Mean Absolute Error")
plt.title("MAE Curve")
plt.legend()

plt.show()
```

# Evaluate the Model on Test Set

```python
# Evaluate on test set
test_loss, test_mae = model.evaluate(X_test_scaled, y_test_scaled)
print(f"Test Loss (MSE): {test_loss:.4f}")
print(f"Test MAE: {test_mae:.4f}")
```

# Make Predictions and Inverse Transform

```python
# Predict on test set
y_pred_scaled = model.predict(X_test_scaled)

# Convert back to original scale
y_pred_original = scaler_y.inverse_transform(y_pred_scaled)
y_test_original = scaler_y.inverse_transform(y_test_scaled)

# Plot Predictions vs True Values
plt.figure(figsize=(10, 5))
plt.plot(y_test_original[:100], label="True Velocity", color='blue')
plt.plot(y_pred_original[:100], label="Predicted Velocity", color='red', linestyle='dashed')
plt.legend()
plt.xlabel("Sample Index")
plt.ylabel("Velocity (original scale)")
plt.title("LSTM Prediction vs True Velocity")
plt.show()
```

# Plot regression

```python
r_value, p_value = pearsonr(y_test_original.flatten(), y_pred_original.flatten())

# Plot
plt.figure(figsize=(7, 6))
sns.regplot(x=y_test_original.flatten(), y=y_pred_original.flatten(), line_kws={"color": "red"})
plt.xlabel("True Velocity")
plt.ylabel("Predicted Velocity")
plt.title(f"Regression Plot\nPearson r = {r_value:.4f}, p = {p_value:.2e}")
plt.grid(True)
plt.tight_layout()
plt.show()
```
